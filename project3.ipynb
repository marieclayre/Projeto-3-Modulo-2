**Projeto 3: Catálogo de Livros com Verificação de Disponibilidade**

Nesse projeto será implementado um catálogo de livros para verificação da disponibilidade. Passos:
➔ Cadastro de Livros:
◆ Os usuários poderão cadastrar livros, incluindo título, autor e número de
exemplares disponíveis.
➔ Validação de Disponibilidade:
◆ A aplicação verificará se o número de exemplares disponíveis é um valor
válido e maior que zero.
➔ Pesquisa de Livros:
◆ Os usuários poderão pesquisar livros por título ou autor.
◆ A aplicação exibirá os resultados da pesquisa e a disponibilidade de cada
livro.

```python
#Class library
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author, copies):
        book = {
            "title": title,
            "author": author,
            "copies": copies
        }
        self.books.append(book)
        print("Livro adicionado com sucesso!")

    def search_by_title(self, search_title):
        results = []
        for book in self.books:
            if search_title.lower() in book["title"].lower() and book["copies"] > 0:
                results.append(book)
        return results

    def search_by_author(self, search_author):
        results = []
        for book in self.books:
            if search_author.lower() in book["author"].lower() and book["copies"] > 0:
                results.append(book)
        return results
```

```python
#Menu principal
def main():
    library = Library()
    while True:
        print("\nSistema de gerenciamento de livros")
        print("1. Adicionar um livro")
        print("2. Procurar livro por livro")
        print("3. Procurar livro por autor")
        print("4. Sair do menu")
        choice = input("Escolha a opção: ")

        if choice == "1":
            title = input("Nome do livro: ")
            author = input("Nome do autor: ")
            copies = int(input("Número de cópias disponíveis: "))
            library.add_book(title, author, copies)

        elif choice == "2":
            search_title = input("Entre com o nome do livro para pesquisar: ")
            books = library.search_by_title(search_title)
            if books:
                print("Livro encontrado!")
                for book in books:
                    print(f"Título: {book['title']}")
                    print(f"Autor: {book['author']}")
                    print(f"Cópias disponíveis: {book['copies']}")
            else:
                print("Livro não encontrado ou cópia não disponível!")

        elif choice == "3":
            search_author = input("Entre com o nome do autor pra pesquisar: ")
            books = library.search_by_author(search_author)
            if books:
                print("Encontrado!")
                for book in books:
                    print(f"Título: {book['title']}")
                    print(f"Autor: {book['author']}")
                    print(f"Cópias disponíveis: {book['copies']}")
            else:
                print("Livro não encontrado ou cópia não disponível!")

        elif choice == "4":
            print("Programa fechando")
            break

        else:
            print("Escolha Inválida. Tente novamente.")


if __name__ == "__main__":
    main()
```
